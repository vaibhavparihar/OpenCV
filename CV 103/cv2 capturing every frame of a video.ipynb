```python
import cv2
import os 

cam = cv2.VideoCapture("komal.mp4")

try:
    if not os.path.exists('pictures'):
        os.makedirs('pictures')
        
except OSError :
    print("Error : creating directory of data")

currentframe = 0
    
while(True):
    ret, frame = cam.read()
    
    if ret :
        name = './pictures/frame' + str(currentframe) + '.jpg'
        print('creating ...' + name)
        
        cv2.imwrite(name,frame)
        
        currentframe += 1
    else :
        break
cam.release()
cv2.destroyAllWindows()
```
